# Making all our features

```python
# Fixing disease binaries on pt table

def fix_binary(df, col_list) :
    for x in col_list:
        df[x] = pd.Series(np.where(df[x]==1, 0, 1))
    return df

bicols = ['ChronicCond_Alzheimer','ChronicCond_Heartfailure','ChronicCond_KidneyDisease','ChronicCond_Cancer','ChronicCond_ObstrPulmonary','ChronicCond_Depression','ChronicCond_Diabetes','ChronicCond_IschemicHeart','ChronicCond_Osteoporasis','ChronicCond_rheumatoidarthritis','ChronicCond_stroke']
pt = fix_binary(pt, bicols)
pt['RenalDiseaseIndicator'] = pd.Series(np.where(pt['RenalDiseaseIndicator']=='Y', 1, 0))
```

```python
# Function to turn date columns to datetime
def col_to_date(df, col_list) :
    for x in col_list:
        df[x] = pd.to_datetime(df[x])
    return df
```

```python
# Make an approx age for each pt on pt table
from datetime import datetime

pt = col_to_date(pt, ['DOB'])

pt['ApproxAge'] = 0
for i in range(len(pt['DOB'])) :
    pt['ApproxAge'][i] = relativedelta(datetime(2009,1,1), pt['DOB'][i]).years
```

```python
# Counting number of diseases on pt table

bicols2 = bicols.copy()
bicols2.append('RenalDiseaseIndicator')

pt['Diseases'] = 0
for i in range(len(pt)) :
    count = 0
    for x in bicols2:
        if pt[x][i] == 1:
            count += 1
    pt['Diseases'][i] = count

display(pt.head())
```

```python
# Make ip_pt table
ip_pt = ip.merge(pt, how='left', on='BeneID')

# Converting date columns to datetime
dtcols_ip_pt = ['ClaimStartDt','ClaimEndDt','AdmissionDt','DischargeDt','DOB']
ip_pt = col_to_date(ip_pt, dtcols_ip_pt)

# Make age column
ip_pt['Age'] = 0
for i in range(len(ip_pt['ClaimStartDt'])) :
    ip_pt['Age'][i] = relativedelta(ip_pt['ClaimStartDt'][i], ip_pt['DOB'][i]).years
```

```python
# Same for OP
op_pt = op.merge(pt, how='left', on='BeneID')

# Converting date columns to datetime
dtcols_op_pt = ['ClaimStartDt','ClaimEndDt','DOB']
op_pt = col_to_date(op_pt, dtcols_op_pt)

# Make age column
op_pt['Age'] = 0
for i in range(len(op_pt['ClaimStartDt'])) :
    op_pt['Age'][i] = relativedelta(op_pt['ClaimStartDt'][i], op_pt['DOB'][i]).years
```

```python
# Total Reimbursement
ip_pt['TotalReimbursement'] = ip_pt['InscClaimAmtReimbursed']+ip_pt['DeductibleAmtPaid']
op_pt['TotalReimbursement'] = op_pt['InscClaimAmtReimbursed']+op_pt['DeductibleAmtPaid']
```

```python
# Adding in fraudulent provider flags!

ip_pt_pro = ip_pt.merge(train, how='left', on='Provider')
op_pt_pro = op_pt.merge(train, how='left', on='Provider')
```

```python
# Writing these tables to processed data

ip_pt_pro.to_csv('../processed_data/ip_pt_pro.csv')
op_pt_pro.to_csv('../processed_data/op_pt_pro.csv')
```

```python
# Making fraudulent and not fraudulent tables

ip_fraud = ip_pt_pro[ip_pt_pro['PotentialFraud']=='Yes']
ip_no_fraud = ip_pt_pro[ip_pt_pro['PotentialFraud']=='No']

op_fraud = op_pt_pro[op_pt_pro['PotentialFraud']=='Yes']
op_no_fraud = op_pt_pro[op_pt_pro['PotentialFraud']=='No']
```

### Resource Usage

```python
# By Age

resource = pt.groupby('ApproxAge')['IPAnnualReimbursementAmt','IPAnnualDeductibleAmt','OPAnnualReimbursementAmt','OPAnnualDeductibleAmt'].sum().reset_index()
resource['IPTotal'] = resource['IPAnnualReimbursementAmt']+resource['IPAnnualDeductibleAmt']
resource['OPTotal'] = resource['OPAnnualReimbursementAmt']+resource['OPAnnualDeductibleAmt']

display(resource.sample(5))

resource.plot(x='ApproxAge', y='IPTotal')
plt.title('IP Total Resource Usage')
plt.show()

resource.plot(x='ApproxAge', y='OPTotal')
plt.title('OP Total Resource Usage')
plt.show()
```

```python
resource2 = pt.groupby('Diseases')['IPAnnualReimbursementAmt','IPAnnualDeductibleAmt','OPAnnualReimbursementAmt','OPAnnualDeductibleAmt'].sum().reset_index()
resource2['IPTotal'] = resource2['IPAnnualReimbursementAmt']+resource2['IPAnnualDeductibleAmt']
resource2['OPTotal'] = resource2['OPAnnualReimbursementAmt']+resource2['OPAnnualDeductibleAmt']

resource2.plot(x='Diseases', y='IPTotal')
plt.title('IP Total Resource Usage by Number of Diseases')
plt.show()

resource2.plot(x='Diseases', y='OPTotal')
plt.title('OP Total Resource Usage by Number of Diseases')
plt.show()
```
